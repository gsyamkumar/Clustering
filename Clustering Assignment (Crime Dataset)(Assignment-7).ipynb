# Hierachical Clustering

```python
# Normalization
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm = norm_func(data.iloc[:,1:])
df_norm
```

```python
#Creating Dendogram using single method
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch#for creating dendogram
z = linkage(df_norm,method='single', metric="euclidean")
plt.figure(figsize=(15,5))
plt.title("Hierarchical Clustering Dendogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,)
plt.show()
```

```python
#it is very difficult to understand and difficult to form clusters by using single method dendogram
#so that, we will try with another method to get better results(i.e,Complete method)
```

```python
#Creating Dendogram using Complete method
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch#for creating dendogram
z = linkage(df_norm,method='complete', metric="euclidean")
plt.figure(figsize=(15,5))
plt.title("Hierarchical Clustering Dendogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,)
plt.show()
```

```python
#here the complete moethod is not advisable 
#so, we are creating the another metohod(i.e, average method)
```

```python
#Creating Dendogram using average method
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch#for creating dendogram
z = linkage(df_norm,method='average', metric="euclidean")
plt.figure(figsize=(15,5))
plt.title("Hierarchical Clustering Dendogram")
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(z,)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.cluster import AgglomerativeClustering
h_complete = AgglomerativeClustering(n_clusters=6, linkage='average',affinity = "euclidean").fit(df_norm)
cluste_labels = pd.Series(h_complete.labels_)
cluster_labels
data['clust']=cluster_labels
data
```

# K-means Clustering

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
kmc = pd.read_csv("crime_data.csv")
kmc
```

```python
#Normalization 
def norm_func(i):
    x =(i-i.min())/(i.max() -i.min())
    return (x)
```

```python
kmc_norm = norm_func(kmc.iloc[:,1:])
kmc_norm
```

```python
#Creating Elbow curve
fig = plt.figure(figsize=(10,8))
WCSS = []
for i in range(1,11):
    clf = KMeans(n_clusters=i)
    clf.fit(kmc_norm)
    WCSS.append(clf.inertia_)#inerita is another name for WCSS and all the calculations are done in background
plt.plot(range(1,11),WCSS)
plt.title("Elbow Curve")
plt.xlabel('WCSS')
plt.ylabel("Number of Clusters")
plt.show()
```

```python
#From the above Elbow curve we can find the 5 clusters
```

```python
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(df_norm)
```

```python
y_kmeans
```

```python
clf.labels_
```

```python
#here we are finding what are the cluster centers
clf.cluster_centers_
```

```python
#used to tell what is the WCSS
clf.inertia_
```

```python
md= pd.Series(y_kmeans)#converting nupy array into pandas series obejct
kmc['clust']=md#creating a new column and assigning it to new column
kmc
```

```python
kmc.iloc[:,1:7].groupby(kmc.clust).mean()
```

```python
kmc.plot(x="Murder", y="Rape", c=clf.labels_,kind="scatter",s=30,cmap=plt.cm.coolwarm)
print("Clusters using K-means")
```

```python
#importing libraries
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
#loading the dataset
db = pd.read_csv("crime_data.csv")
db
```

```python
db.info()
```

```python
dbs = db.iloc[:,1:]
dbs
```

```python
array = dbs.values
array
```

```python
#Converting the data into z-scores
stscaler = StandardScaler().fit(array)
X =stscaler.transform(array)
X
```

```python
dbscan = DBSCAN(eps=0.475, min_samples=5)
dbscan.fit(X)
```

```python
#here -1 imples the noise values(radius)
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])  
cl
```

```python
pd.set_option("display.max_rows", None)  
cl
```

```python
db = pd.concat([dbs,cl],axis=1)
db
```
